```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets

ufos = pd.read_csv("ufo-scrubbed-geocoded-time-standardized-00.csv",
                  names = ["date", "city", "state", "country",
                          "shape", "duration_seconds", "duration",
                          "comment", "report_date", "latitude", "longitude"],
                  parse_dates = ["date", "report_date"])
```

```python
ufos
```

Fill the empty values.

```python
ufos = ufos.fillna({'country':'unknown', 'state': 'unknown', 'shape':'unknown'})
```

Exclude the column 'comment', keep the rest of the columns.

```python
ufos_col = ufos[["date", "city", "state", "country",
                          "shape", "duration_seconds", "duration",
                          "report_date", "latitude", "longitude"]]
ufos_col
```

Set the numerical columns as "duration_seconds", "latitude" and "longitude".

```python
ufos_numerical = ufos[["duration_seconds", "latitude", "longitude"]]
ufos_numerical
```

## Question 1:

User can change the x and y fields using the columns of ufos dataset.

Also, user can change colormap using values in the numerical columns.

I would like the users to select any 2 columns since the plot presented has a certain kind of meaning.

```python
@ipywidgets.interact(col_name1 = ufos.columns, col_name2 = ufos.columns, colormap = ufos_numerical.columns)
def create_plot(col_name1, col_name2, colormap):
    fig,ax = plt.subplots(figsize=(10,6))
    ax.scatter(ufos[col_name1], ufos[col_name2], c=np.log10(ufos[colormap]))
    ax.set_xlabel(col_name1)
    ax.set_ylabel(col_name2)
    plt.show()
```

```python
ufos_categorical = ufos[["date", "city", "state", "country", "shape", "report_date"]]
ufos_categorical
```

## Question 2:

Create a histogram which can allow users to select x field and the bins. 

I set the bins range from 1 to 50 with a step of 1.

Users can select from "date", "city", "state", "country", "shape", "report_date" as the x field, because the plot has real-world meaning by using these columns.

```python
bins_slider = ipywidgets.IntSlider(min=1,max=50,step=1)
@ipywidgets.interact(column = ufos_categorical.columns, bins=bins_slider)
def creat_hist(column, bins):
    fig,ax = plt.subplots(figsize=(10,6))
    ax.hist(x = ufos[column],bins = bins)
    ax.set_xlabel(column)
    plt.show()
```
